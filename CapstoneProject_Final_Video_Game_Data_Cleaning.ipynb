```python
import pandas as pd
import numpy as np
```

```python
df_rev = pd.read_json('Video_Games.json.gz', orient='columns',lines=True,nrows=200000)
df_rev["reviewTime"] = pd.to_datetime(df_rev["reviewTime"])
df_rev.head()
```

```python
import re
def cleaning(text):
    text = re.sub('[^0-9A-Za-z]+',' ',text)
    text= re.sub('(?<!\w)\d+','',text)
    text=re.sub('-(?!\w)|(?<!\w)-','',text)
    text=text.lower()
    text=re.sub(' +',' ',text)
    return text
df_rev["reviewText"]=df_rev["reviewText"].astype(str)
df_rev["summary"]=df_rev["summary"].astype(str)
df_rev["reviewText"]=df_rev["reviewText"].apply(cleaning)
df_rev["summary"]=df_rev["summary"].apply(cleaning)
```

```python
df_rev.head()
```

```python
df_rev["review"] = df_rev['reviewText'].astype(str) +" "+ df_rev["summary"]
df_rev=df_rev.drop(['unixReviewTime','vote','style','image'],axis=1)
df_rev
```

```python
df_rev["Number_of_Words_in_review"] = df_rev["review"].apply(lambda n: len(n.split()))

df_rev
```

```python
df_rev.isnull().sum()
```

```python
df_rev["reviewerName"].fillna(df_rev["reviewerName"].mode()[0], inplace=True)
```

```python
df_rev.isnull().sum()
```

```python
df_rev['Sentiment']=df_rev['overall']
df_rev['Sentiment'] = df_rev['Sentiment'].replace(5,'positive')
df_rev['Sentiment'] = df_rev['Sentiment'].replace(4,'positive')

df_rev['Sentiment'] = df_rev['Sentiment'].replace(2,'negetive')
df_rev['Sentiment'] = df_rev['Sentiment'].replace(1,'negetive')

df_rev['Sentiment'] = df_rev['Sentiment'].replace(3,'neutral')
```

```python
df_rev
```

# Meta

```python
df_meta = pd.read_json('meta_Video_Games.json.gz', orient='columns',lines=True)
```

```python
df_meta['category']=df_meta['category'].apply(lambda x: ', '.join(map(str, x)))
df_meta['description']=df_meta['description'].apply(lambda x: ', '.join(map(str, x)))
df_meta['also_buy']=df_meta['also_buy'].apply(lambda x: ', '.join(map(str, x)))
df_meta['feature']=df_meta['feature'].apply(lambda x: ', '.join(map(str, x)))
df_meta['rank']=df_meta['rank'].apply(lambda x: ', '.join(map(str, x)))
df_meta['also_view']=df_meta['also_view'].apply(lambda x: ', '.join(map(str, x)))
df_meta['imageURL']=df_meta['imageURL'].apply(lambda x: ', '.join(map(str, x)))
```

```python
import re
def cleaning(text):
    text = re.sub('[^0-9A-Za-z]+',' ',text)
    text = re.sub('span',' ',text)
    text = re.sub('\n',' ',text)
    text = text.lower()
    return text

df_meta['category']=df_meta['category'].apply(cleaning)
df_meta['description']=df_meta['description'].apply(cleaning)
```

```python
import re
def cleaning(text):
    text = re.sub('[^0-9A-Za-z]+',' ',text)
    text = re.sub('by',' ',text)    
    text = text.lower()
    return text
df_meta['brand']=df_meta['brand'].apply(cleaning)
```

```python
df_meta['rank'][43]
```

```python
import re
def cleaning(text):
    text = text.lower()
    text = re.sub(r'\si.*',' ',text)
    text = re.sub('>#',' ',text) 
    text = re.sub(r':.*',' ',text)
    text = re.sub(r'\ss.*',' ',text)
    text = re.sub(r'\sp.*',' ',text)
    text = re.sub(r'[\sa-z.*]',' ',text)
    text = re.sub(',',' ',text)
    text = re.sub(r"\s", "", text)
    
    return text
df_meta['rank']=df_meta['rank'].apply(cleaning)
```

```python
df_meta.drop(columns=['tech1','fit','tech2','similar_item','date','details','imageURLHighRes','also_buy','also_view','feature','price'],inplace=True)
```

```python
df_meta=df_meta.mask(df_meta == '')
```

```python
df_meta['rank'][52242]
```

```python
#df_meta['rank'] = df_meta['rank'].astype(int)
df_meta['rank'] = pd.to_numeric(df_meta['rank'])
```

```python
df_meta.isnull().sum()
```

```python
df_meta.dtypes
```

```python
df_meta["category"].fillna(df_meta["category"].mode()[0], inplace=True)
df_meta["description"].fillna(df_meta["description"].mode()[0], inplace=True)
df_meta["brand"].fillna(df_meta["brand"].mode()[0], inplace=True)
df_meta["rank"].fillna(df_meta["rank"].mean(), inplace=True)
df_meta["title"].fillna(df_meta["title"].mode()[0], inplace=True)
df_meta["main_cat"].fillna(df_meta["main_cat"].mode()[0], inplace=True)

```

```python
df_meta.isnull().sum()
```

```python
df_meta['image_uploaded']=np.nan
df_meta['imageURL']=df_meta['imageURL'].astype(str)
a=0

for i in df_meta['imageURL']:
    z=re.match("^http",i, re.IGNORECASE)
    if bool(z) == True:
        df_meta['image_uploaded'][a]='yes'
    else:
        df_meta['image_uploaded'][a]='no'
    a=a+1
```

```python
df_meta.drop(columns=['imageURL'],inplace=True)
```

```python
merged = pd.merge(df_rev,df_meta,on=['asin'], how='inner')
merged.shape
```

```python
merged.drop_duplicates(subset=["reviewText","summary"],keep='first', inplace=True)
```
